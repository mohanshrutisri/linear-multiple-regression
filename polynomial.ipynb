```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=6*np.random.rand(200,1)-3
y=0.8*x**2+0.9*x+2+np.random.randn(200,1)
```

```python
plt.plot (x,"y")
plt.plot(y,"r")
```

```python
plt.scatter(x,y)
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(x,y,"r")
plt.plot(x_train,lr.predict(x_train),"b")
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
x_train_tran = poly.fit_transform(x_train)
x_test_tran = poly.fit_transform(x_test)
print(x_train[0])
print(x_test_tran[0])


```

```python
model = LinearRegression()
model.fit(x_train_tran,y_train)
```

```python
yy_pred = model.predict(x_test_tran)
r2_score(y_test,yy_pred)
```

```python
plt.scatter(x,y)
plt.plot(x_train,lr.predict(x_train),"b")
plt.plot(x_train_tran,model.predict(x_train_tran),"r")
```
